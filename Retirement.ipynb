## Out Data Preprocessing

```python
raw_out_data
```

```python
out_birth = pd.DataFrame(raw_out_data.iloc[2:, [2, 6]])
```

```python
out_birth.columns = ['생일', '퇴직일자']
```

```python
birth_list2 = []
for i in out_birth['생일']:
    birth_year = int(i[0:4])
    age = 2018-birth_year+1
    birth_list2.append(age)
```

```python
out_birth['나이'] = birth_list2
```

```python
bins = [x for x in range(-1, 121, 10)]
labels = [x for x in range(0, 12)]
cut = pd.cut(out_birth['나이'], labels=labels, bins=bins, include_lowest=True)
out_birth['cut']=cut
```

```python
out_birth = out_birth.dropna()
```

```python
out_birth.isnull().sum()
```

```python
out_birth['퇴직일자'] = out_birth['퇴직일자'].astype(int)
```

```python
out_birth_retire=out_birth.loc[out_birth['퇴직일자']>=20150630, :]
```
